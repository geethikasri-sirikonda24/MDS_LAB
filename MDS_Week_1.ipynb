```python
import time
import numpy as np

# Define the size of the arrays/lists
size = 1000000

# --- Python List Performance ---
start_time = time.time()
python_list_a = [i for i in range(size)]
python_list_b = [i for i in range(size)]
python_list_sum = [a + b for a, b in zip(python_list_a, python_list_b)]
end_time = time.time()
python_list_time = end_time - start_time
print(f"Time taken for Python list: {python_list_time:.6f} seconds")

# --- NumPy Array Performance ---
start_time = time.time()
numpy_array_a = np.arange(size)
numpy_array_b = np.arange(size)
numpy_array_sum = numpy_array_a + numpy_array_b
end_time = time.time()
numpy_array_time = end_time - start_time
print(f"Time taken for NumPy array: {numpy_array_time:.6f} seconds")

print(f"\nNumPy is approximately {python_list_time / numpy_array_time:.2f} times faster for this operation.")
```

### Memory Comparison: Python Lists vs. NumPy Arrays

Let's compare the memory consumption for Python lists and NumPy arrays for similar data sizes.

```python
import sys

# Assuming 'size' is already defined from the previous cell

# --- Python List Memory Usage ---
python_list = [i for i in range(size)]
# For a Python list, we need to consider the size of the list object itself
# plus the size of each element within the list.
# sys.getsizeof(list_obj) gives the size of the list object, not its contents.
# For integers, sys.getsizeof(0) gives the base size of an int object.
# python_list_memory = sys.getsizeof(python_list) + sum(sys.getsizeof(i) for i in python_list[:1000]) * (size // 1000) # Estimate for large list
# A more accurate estimation for memory for a list of integers would be:
# python_list_memory = sys.getsizeof(python_list) # Size of the list structure
# python_list_memory += sum(sys.getsizeof(x) for x in python_list) # Size of individual elements (can be very slow for large lists)
# Let's use a more practical approximation for demonstration:
python_list_memory_approx = sys.getsizeof(python_list) + size * sys.getsizeof(python_list[0]) # List object + average size of elements

print(f"Memory taken for Python list (approximate): {python_list_memory_approx / (1024*1024):.4f} MB")

# --- NumPy Array Memory Usage ---
numpy_array = np.arange(size)
# NumPy arrays store data contiguously and efficiently.
# nbytes gives the total bytes consumed by the array's data.
numpy_array_memory = numpy_array.nbytes
print(f"Memory taken for NumPy array: {numpy_array_memory / (1024*1024):.4f} MB")

print(f"\nNumPy array uses approximately {python_list_memory_approx / numpy_array_memory:.2f} times more memory than a Python list for this operation.")
```
